# Problem 2 - N-th Fibonacci number

### Python

````python
def fibonacci_number(n: int) -> int:
    f = [0, 1]
    while len(f) <= n:
        f.append(f[-1] + f[-2])
    
    return(f[n])


n1 = 7
n2 = 8

result_1 = fibonacci_number(n1)
result_2 = fibonacci_number(n2)

print(result_1)
print(result_2)
````

### C++
```C++
#include <iostream>
#include <vector>

using namespace std;

int main() {
    int n;
    cin >> n;
    vector<int64_t> f = {0, 1};
    
    while ((int)f.size() <= n) {
        f.push_back(f[f.size() - 1] + f[f.size() - 2]);
    }
    
    cout << f[n];
}
```
